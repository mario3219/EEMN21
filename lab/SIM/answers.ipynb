# Q7 Why is the pressure gradient steeper in the main channel as compared to the central inlet?



My guess: The path from central inlet to main is rather small, where in the intersection all other inlets meet, summarizing pressures. The main channel travels in one geometry to zero pressure, therefore a steeper drop.

In conclusion: the central inlet ends at non-zero pressure, main ends at zero pressure.
